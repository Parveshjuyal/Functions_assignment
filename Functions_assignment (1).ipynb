# 7 What is the difference between”==” and “is” operators in

      a =[1,2,3]
      b =[1,2,3]
      print(a == b)# ✅ True (same values)

   # “is”

    - Checks if two variables point to the exact same object
      in memory.

    - a =[1,2,3]
      b = a # b references the SAME list as a

    - To check if two variables reference the same object

# 10 What is the difference between implicit and explicit type casting?

  # Implicit Type Casting
     Automatic (Python’s choice).
     Python automatically converts one type to another
     without your input.

  # Integer + Float = Float (Python auto-converts)  
     result =5+2.5# Python makes 5 → 5.0  
     print(result)# Output: 7.5  
  
  # Explicit Type Casting
     Manual (Your choice).
     You forcefully convert a type using functions like   int (), str(), etc.
  # String number → Real number  
     text_age ="25"
     real_age =int(text_age)# YOU convert it  
     print(real_age +5)# Output: 30
